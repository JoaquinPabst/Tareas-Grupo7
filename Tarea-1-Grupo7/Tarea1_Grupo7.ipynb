## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%%sql
CREATE TABLE Usuarios (
    ID_usuario INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    apellido VARCHAR (45),
    correo VARCHAR (45) NOT NULL,
    contraseña VARCHAR (50) NOT NULL,
    fecha_registro TIMESTAMP NOT NULL,

    PRIMARY KEY (ID_usuario)
);
```

### Tabla "Creadores"

```python
%%sql
CREATE TABLE Creadores (
    ID_creador INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    apellido VARCHAR (45) NOT NULL,
    popularidad INTEGER,
    Estilo VARCHAR (45),

    PRIMARY KEY (ID_creador)
);
```

### Tabla "Tipos_NTF"

```python
%%sql
CREATE TABLE Tipos_NFT (
    ID_tipo INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (20) NOT NULL,

    PRIMARY KEY (ID_tipo)
);
```

### Tabla "NFT"

```python
%%sql
CREATE TABLE NFT (
    ID_nft INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    ID_creador INTEGER NOT NULL,
    id_tipo INTEGER,

    PRIMARY KEY (ID_nft),
    FOREIGN KEY (ID_creador)
        REFERENCES Creadores (ID_creador),
    FOREIGN KEY (id_tipo)
        REFERENCES Tipos_NFT (ID_tipo)
);
```

### Tabla "Precio NFT"

```python
%%sql
CREATE TABLE Precio_NFT (
    ID_nft INTEGER,
    fecha TIMESTAMP NOT NULL,
    precio INT NOT NULL,

    PRIMARY KEY (ID_nft, fecha),
    FOREIGN KEY (ID_nft)
        REFERENCES NFT (ID_nft)
);
```

### Tabla "Usuario tiene NFT"

```python
%%sql
CREATE TABLE Usuario_tiene_NFT (
    id_usuario INTEGER,
    id_nft INTEGER,
    PRIMARY KEY (id_usuario, id_nft),
    FOREIGN KEY (id_usuario)
        REFERENCES Usuarios (ID_usuario),
    FOREIGN KEY (id_nft)
        REFERENCES NFT (ID_nft)        
);
```

### Consulta 7
**Top 4 NFTs más caros actualmente de un id_tipo X .**

Ejemplo:

|NFT | 
|--------|
|Minecraft Sphere | 
|The Binary Girl |
|Koalas in Love | 
|Art Dodger | 

**Solución:** Usamos SELECT DISTINCT para mostrar el nombre de la tabla que haremos en FROM

En el FROM havemos un triple INNER JOIN, la primera tabla (info_precio) es una reduccion de precio_nft en la que seleccionamos solo fecha y precio

La segunda (fecha_actual) es una en la que seleccionamos el id_nft y la fecha maxima del nft (mas actual), esto de la tabla precio_nft, la agrupamos  por id. La primera y la segunda tabla las vinculamos por la fecha. (asi nos deja una tabla solo con las fechas actuales)

La tercera tabla (info_nft) es una reduccion de nft en la cual seleccionamos id_nft y su nombre, luego la vinculamos a la primera por id_nft

Con esto creamos una tabla en la que tenemos los datos del nft y el precio que tienen en la fecha actual

agrupamos por nombre (inf_nft.nombre) ordenamos de forma decendiente por el precio maximo y limitamos a cuatro, de esta forma obtenemos el top 4 actual.

```python
%%sql
SELECT 
	info_nft.nombre
FROM (
		SELECT 
			fecha, 
			precio
		FROM 
			precio_nft
	) info_precio
	INNER JOIN (
		SELECT 
			id_nft, 
			MAX(fecha) AS max_fecha
		FROM 
			precio_nft
		GROUP BY 
			id_nft
	) fecha_actual
	ON
		info_precio.fecha = fecha_actual.max_fecha
	INNER JOIN (
		SELECT 
			id_nft, 
			nombre
		FROM 
			nft 
	) info_nft
ON 
	fecha_actual.id_nft = info_nft.id_nft
GROUP BY 
	info_nft.nombre
ORDER BY 
	MAX(precio) DESC
LIMIT 4;
```

### Consulta 10  
**El NFT con mayor precio actual, que pertenezca a un creador X.**

Ejemplo:
   
|NFT |Valor|
|-|-------|
|Minecraft Sphere |18000|

**Solución:** Con SELECT mostramos el nobre del nft y su precio

En el FROM uniremos 5 tablas, la primera (info_precio) esa una reduccion de precio_nft, en la que seleccionamos fecha y precio

La unimos con una reduccione de precio_nft (fecha_actual), en la que msotramos el id del nft y su fecha actual, la unimos con la anterior por su fecha, asi nos deja una tabla solo con las fechas actuales

Luego unimos una tercera tabla (info_nft), la que es una reduccion de nft la cual nos muestra el id y el nombre, la unimos a la anterior por el id_nft

Luego la unimos con usuario_tiene_nft, por el id_nft

y finalmente unimos una reduccion de usuario en la cual mostramos su nombre e id_usuario, usamos WHERE para seleccionar al usuario que queremos conozer su nft mas alto (X), esta la unimos a usuario_tiene_nft por el id_usuario


Agrupamos por info_nft.nombre y info_precio.precio, ordenamos por precio maximo de forma decendiente, para dejar el nft mas caro arriba y luego limitamos a uno.

```python
%%sql
SELECT 
    info_nft.nombre AS "NFT",
    info_precio.precio AS "Valor"
FROM (
        SELECT 
            fecha, 
            precio
        FROM 
            precio_nft
    ) info_precio
    INNER JOIN (
        SELECT 
            id_nft, 
            MAX(fecha) AS max_fecha
        FROM 
            precio_nft
        GROUP BY 
            id_nft
    ) fecha_actual
    ON
        info_precio.fecha = fecha_actual.max_fecha
    INNER JOIN (
        SELECT 
            id_nft, 
            nombre,
            id_creador
        FROM 
            nft 
    ) info_nft
    ON 
        fecha_actual.id_nft = info_nft.id_nft
    INNER JOIN (
        SELECT
            id_creador,
            nombre
        FROM 
            creadores
        WHERE
            id_creador = 1 /*Cambiar a eleccion*/
    ) info_creador
    ON info_creador.id_creador = info_nft.id_creador

GROUP BY
    info_nft.nombre,
    info_precio.precio
ORDER BY 
    MAX(precio) DESC
LIMIT 1;
```
